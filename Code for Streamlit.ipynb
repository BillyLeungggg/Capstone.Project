```python
df["tags"] = df["product_description"] + " " + df["category_name"] + " " + df["category_description"] + " " + df["subcategory_name"] + " " + df["subcategory_description"]
```

```python
df["tags"] = df['tags'].str.replace('&', 'and', regex=False)
df["tags"] = df['tags'].str.replace('.', '', regex=False)
df["tags"] = df['tags'].str.replace('\n', ' ', regex=False)
df["tags"] = df['tags'].str.replace('(', '', regex=False)
df["tags"] = df['tags'].str.replace(')', '', regex=False)
df["tags"] = df['tags'].str.replace('none', '', regex=False)
```

# Feature Selection


```python
new_df = df[["unique_id", "product_name","tags"]]
```

```python
new_df
```

```python
new_df.shape
```

Put texts into vector

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=823, stop_words="english")
```

```python
cv
```

```python
vector = cv.fit_transform(new_df["tags"].values.astype("U")).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```
