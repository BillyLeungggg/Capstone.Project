```python
# Sort the categories from highest to lowest
counts = df['category_name'].value_counts()
sorted_categories = counts.index

# Create the count plot
plt.figure(figsize=(8, 6))  # Set the figure size
sns.countplot(x='category_name', data=df, order=sorted_categories)  # Set the x-axis to the 'category_name' column, sorted from high to low
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels by 45 degrees and align them to the right
plt.title('Count of Products by Category')  # Set the title
plt.xlabel('Category Name')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.show()  # Display the plot
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['price(Rs.)'], bins=16, edgecolor='black')
plt.xlabel('Price (Rs.)')
plt.ylabel('Count')
plt.title('Price Distribution')
plt.show()
```

Histogram: About 600 products's price is within 500 dollars.


```python
# Heatmap

pivot_table = pd.pivot_table(df, values='price(Rs.)', index='category_name', aggfunc='mean')
pivot_table = pivot_table.sort_values(by='price(Rs.)', ascending=False)

plt.figure(figsize=(11, 11))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.2f')
plt.xlabel('Category Name')
plt.ylabel('Price (Rs.)')
plt.title('Price Heatmap')
plt.show()
```

Heatmap: used the pd.pivot_table() function to calculate the average price for each category and created a pivot table. Then, used the seaborn library's heatmap() function to create the heatmap, with the category names on y axis, prices on the x axis.
